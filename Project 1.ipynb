```python
missing_values={}
columns=[]
no_missing_values=[]
for cols in data.columns:
    ma=data[cols].isna().sum()
    columns.append(cols)
    no_missing_values.append(ma)
missing_values["Columns"]=columns
missing_values["Number of Missing Values"]=no_missing_values
#print(missing_values)
missing_values_df=pd.DataFrame.from_dict(missing_values) #https://stackoverflow.com/questions/4527454/python-dictionary-contains-list-as-value-how-to-update
missing_values_df
```

```python
numeric_col=data.columns[5:len(data.columns)+1]
```

```python
x_m=[]
x_f=[]
for c in numeric_col:
    if not "95%" in c :
        if "Male" in c:
            x_m.append(c)
        else:
            x_f.append(c)
    
#x_m
#x_f
#x_m.append("Local Authority")
print(x_f)
```

```python
#https://www.shanelynn.ie/select-pandas-dataframe-rows-and-columns-using-iloc-loc-and-ix/#iloc-selection
y=data.loc[data["Ward name"]=="Village",x_m] 
z=data.loc[data["Ward name"]=="Village",x_f]
y.iloc[[0]]
z.iloc[[0]]
```
